#### Encoding
In Task 1 hebben we al gekeken naar de categorische variabelen. Laten we die variabelen maar eens enconden.

```python
lijst
```

```python
def encode(df):
    
    set_sex = {'F':0, 'M':1}
    df['sex'] = df['sex'].map(set_sex)

    set_reason = {'course': 0, 'other': 1, 'home': 2, 'reputation' : 3}
    df['reason'] = df['reason'].map(set_reason)
    return df
    
```

### Outliers
in de basis verkennende analyse hebben we gekeken naar de outliers. hieronder staan ze:




```python
def check_outliers():
    # Maak aparte boxplots voor elke kolom.
    fig, axes = plt.subplots(2, 4, figsize=(15, 8))

    # Boxplot voor 'Sex'
    sns.boxplot(ax=axes[0, 0], x='sex', data=student_subset)
    axes[0, 0].set_title('Boxplot of sex')

    # Boxplot voor 'Medu'
    sns.boxplot(ax=axes[0, 1], x='Medu', data=student_subset)
    axes[0, 1].set_title('Boxplot of Medu')

    # Boxplot voor 'Fedu'
    sns.boxplot(ax=axes[0, 2], x='Fedu', data=student_subset)
    axes[0, 2].set_title('Boxplot of Fedu')

    # Boxplot voor 'reason'
    sns.boxplot(ax=axes[0, 3], x='reason', data=student_subset)
    axes[0, 3].set_title('Boxplot of reason')

    # Boxplot voor 'age'
    sns.boxplot(ax=axes[1, 0], x='age', data=student_subset)
    axes[1, 0].set_title('Boxplot of age')

    # Boxplot voor 'absences'
    sns.boxplot(ax=axes[1, 1], x='absences', data=student_subset)
    axes[1, 1].set_title('Boxplot of absences')

    # Boxplot voor 'G1'
    sns.boxplot(ax=axes[1, 2], x='G1', data=student_subset)
    axes[1, 2].set_title('Boxplot of G1')

    # Boxplot voor 'G3'
    sns.boxplot(ax=axes[1, 3], x='G3', data=student_subset)
    axes[1, 3].set_title('Boxplot of G3')

    plt.tight_layout()
    plt.show()
```

```python
check_outliers()
```

Laten we nu de outliers verwijderen

```python
def verwijder_outliers(series):
    # Removing outliers based on IQR
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1

    # Using the IQR filter to remove outliers
    return series.loc[(series >= Q1 - 1.5 * IQR) & (series <= Q3 + 1.5 * IQR)]
```

```python
all_vars = cat_vars + num_vars

for i in all_vars:
    student_subset[i] = verwijder_outliers(student_subset[i])

```

```python
student_subset = student_subset.dropna()
```

```python
check_outliers()
```

#### multicollinearity
Laten we eerst kijken naar de multicollinearity. Multicollineariteit is een statistisch fenomeen waarin twee of meer onafhankelijke variabelen in een regressiemodel sterk gecorreleerd zijn

```python
vif_data = pd.DataFrame()
vif_data["Variable"] = X_train_simple.columns

# Calculate VIF for each variable
vif_data["VIF"] = [variance_inflation_factor(X_train_simple.values, i) for i in range(X_train_simple.shape[1])]

print(vif_data)
```


De Variance Inflation Factor (VIF) meet hoe sterk variabelen in een regressiemodel met elkaar gecorreleerd zijn. Hier is een beknopte uitleg van de VIF-waarden voor de variabelen:
- age (VIF: 20.2): Hoge multicollineariteit,kan problematisch zijn.
- sex (VIF: 1.8): Lage multicollineariteit, gunstig.
- Medu (VIF: 6.8): Hoge multicollineariteit, kan problematisch zijn.
- reason (VIF: 2.2): Lage multicollineariteit, positief.
- absences (VIF: 2.1): Lage multicollineariteit, gunstig.

Kort samengevat:

Lage VIF (rond 1) is goed.
Matige VIF (5-10) kan aandacht vereisen.
Hoge VIF (>10) kan problematisch zijn, overweeg aanpassingen.
Voor variabelen met hoge VIF (age en G1u)heb ik gekeken naar de variabelen en ik heb gekozen om age eruit te gooien.

### Forward Selection
Nu gaan we via forward selection kijken naar het model om er achter te komen wat de hoogste accuracy is.

```python
# Deze variabelen bestaan niet meer
all_vars.remove('age')
all_vars.remove('G3')
all_vars.remove('Fedu')
```

```python
highest = 0
best_variable = ""
regression_model = LinearRegression()
# Iterate through all possible combinations
for r in range(1, len(all_vars) + 1):
    for variable_combination in combinations(all_vars, r):
        # Fit a linear regression model
        X = X_train_simple.iloc[:, [X_train_simple.columns.get_loc(var) for var in variable_combination]]
        X_test = X_test_simple.iloc[:, [X_test_simple.columns.get_loc(var) for var in variable_combination]]

        regression_model.fit(X, y_train_simple)

        # Make predictions
        y_predict = regression_model.predict(X_test)

        # Calculate R-squared score
        r_squared = r2_score(y_test_simple, y_predict)

        # Update best_variable if R-squared is improved
        if highest < r_squared:
            highest = r_squared
            best_variable = f"Variables: {', '.join(variable_combination)} | R_squared: {r_squared}"

print(best_variable)




```

uit dit stuk code is gebleken dat alle overgebleven variabelen (Medu, reason, absences, G1), de hoogste R_squared geeft. R_squared wil zeggen: hoger bij de 1, hoe hoger de accuracy. Laten we dit dus trainen.

## andere modellen



```python
models = {
    'Linear Regression': (LinearRegression(), {}),
    'Ridge Regression': (Ridge(), {'alpha': [0.1, 1.0, 10.0, 0.2, 0.3, 0.4, 0.5]}),
    'Lasso Regression': (Lasso(), {'alpha': [0.1, 1.0, 10.0, 0.2, 0.3, 0.4, 0.5]}),
    'Random Forest Regression': (RandomForestRegressor(), {'n_estimators': [50, 100, 200]}),
    'Polynomial Regression': (make_pipeline(PolynomialFeatures(), LinearRegression()), {'polynomialfeatures__degree': [2, 3, 4]}),
    'Elastic Net Regression': (ElasticNet(), {'alpha': [0.1, 1.0, 10.0], 'l1_ratio': [0.1, 0.5, 0.9]}),
}

# Perform grid search for each model
results = {}
for model_name, (model, param_grid) in models.items():
    grid_search = GridSearchCV(model, param_grid, scoring='neg_mean_squared_error', cv=5)
    grid_search.fit(X_train_simple, y_train_simple)
    
    # Get the best model
    best_model = grid_search.best_estimator_
    
    # Evaluate the best model on the test set
    y_pred = best_model.predict(X_test_simple)
    mse = mean_squared_error(y_test_simple, y_pred)
    
    r_squared = r2_score(y_test_simple, y_predict)
    
    # Store results
    results[model_name] = {
        'Best Model': best_model,
        'Best Parameters': grid_search.best_params_,
        'Mean Squared Error': mse,
        'R_squared': r_squared
    }

# Print the results
for model_name, result in results.items():
    print(f"Results for {model_name}:")
    print(f"  Best Model: {result['Best Model']}")
    print(f"  Best Parameters: {result['Best Parameters']}")
    print(f"  Mean Squared Error on Test Set: {result['Mean Squared Error']}")
    print(f"  R2 score {result['R_squared']}")
    print("\n")
```

Je ziet dat alle modellen een lage MSE hebben, laten we dan maar voor de Linear regression kiezen, deze is immers het laagst
